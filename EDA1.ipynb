```python
df.drop(['Area Locality','Floor','Posted On'], axis=1, inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
cols = ['Area Type', 'City','Furnishing Status','Tenant Preferred','Point of Contact',]

df[cols] = df[cols].apply(LabelEncoder().fit_transform)

print(df['Area Type'].unique())
print(df['City'].unique())
print(df['Furnishing Status'].unique())
print(df['Tenant Preferred'].unique())
print(df['Point of Contact'].unique())

```

```python
plt.figure(figsize = (12,8))
sns.heatmap(df.corr() , annot = True , cmap = "YlGnBu")
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = df.drop('Rent',axis=1).values
y = df['Rent'].values
```

```python
x_train ,x_test ,y_train ,y_test=train_test_split(x,y ,test_size=0.2,random_state=42)
```

```python
x_train.shape
```

```python
x_test.shape
```

MODELING

```python
from sklearn.preprocessing import StandardScaler
print(x_test)
```

```python
ss=StandardScaler()
x_train =ss.fit_transform(x_train)
x_test=ss.fit_transform(x_test)
```

```python
print(x_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
random_forest=RandomForestRegressor()
random_forest.fit(x_train , y_train)
print("Random Forest Training Accuracy:", str((random_forest.score(x_train , y_train))*100)+'%')
print("Random Forest Testing Accuracy:", str((random_forest.score(x_test , y_test))*100)+'%')
```
