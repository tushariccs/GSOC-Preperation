# #Univariate Analysis


```python
sal.boxplot()
```

```python
sns.displot(sal,x="Salary",hue="Years of Experience")
plt.tight_layout()
plt.show()
```

```python
sns.heatmap(sal.corr())
plt.show()
```

# Bivariate Analysis
# Relational Plot

```python
sns.relplot(x="Years of Experience",y="Salary",hue="Gender",data=sal)
plt.tight_layout()
plt.show()
```

```python
sns.histplot(data=sal,x="Years of Experience",y="Salary",hue="Gender",)
plt.show()
```

```python
sns.scatterplot(data=sal,x="Years of Experience",y="Salary",hue="Gender")
plt.tight_layout()
plt.show()
```

```python
sns.lineplot(data=sal,x="Years of Experience",y="Salary",hue="Gender")
plt.tight_layout()
plt.show()
```

```python
sal.groupby("Gender")[["Salary"]].mean().T.plot(kind='barh', figsize=(10,10))
plt.show()
```

```python
sns.barplot(data=sal,x="Years of Experience",y="Salary",hue="Gender")
plt.tight_layout()
plt.show()
```
